## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Remove duplicate rows
apps_df = apps_df.drop_duplicates()

# Cleaning Installs column
apps_df['Installs'] = apps_df['Installs'].str.replace('[+,]', '', regex=True)
apps_df['Installs'] = pd.to_numeric(apps_df['Installs'], errors='coerce')

# Cleaning Price column
apps_df['Price'] = apps_df['Price'].str.replace('$', '', regex=False)
apps_df['Price'] = pd.to_numeric(apps_df['Price'], errors='coerce')

# Handling Size column
apps_df['Size'] = apps_df['Size'].replace('Varies with device', np.nan)
apps_df['Size'] = apps_df['Size'].str.replace('M', '')
apps_df['Size'] = pd.to_numeric(apps_df['Size'], errors='coerce')

# Handling missing ratings
apps_df['Rating'].fillna(apps_df['Rating'].mean(), inplace=True)

# Convert Last Updated to datetime
apps_df['Last Updated'] = pd.to_datetime(apps_df['Last Updated'], errors='coerce')
````

### What all manipulations have you done and insights you found?

Answer Here.
The following data wrangling steps were performed to prepare the dataset for analysis:

Removed duplicate records to avoid biased results.

Cleaned the Installs column by removing special characters and converting it to numeric format.

Cleaned the Price column by removing currency symbols and converting it to float.

Converted Size into numeric values and handled “Varies with device” as missing data.

Filled missing Rating values using the mean rating.

Converted Last Updated column to datetime format for time-based analysis.

Insights Found:

Free apps dominate the Play Store ecosystem.

Apps with smaller sizes tend to have higher installs.

Ratings are generally high, but engagement varies significantly across categories.

Price negatively impacts the number of installs.
